### Raising an Exception

##### There are scenario where we might want to stop our program by raising an exception if a condition occurs. We can do this using rasie keyword

```python
number= 20
if number > 10:
    raise Exception(f"The number is greater than 10. number :: {number}")
print(number)
```

### Debugging using Assert

##### assert-keyword is used to check the condition is met and let Python raise the AssertionError if condition isn't meet.

```python
number= 20

assert(number <10), f"The number :: {number} is not less than 10"
```

### Proceeding with successful Try With else

##### else statement will execute the block of code only when there is no exception

```python
def linux_sys():
    import sys
    if not sys.platform.startswith("linux"):
        raise RuntimeError(f"Func run on linux system")
    print("Linux Thing")

try:
    linux_sys()
except RuntimeError as err:
    print(err)
    print("This is not a linux sys")
else:
    print("Doing more linux thing")
```

```python
def linux_sys():
    import sys
    if not sys.platform.startswith("linux"):
        raise RuntimeError(f"Func run on linux system")
    print("Linux Thing")

try:
    linux_sys()
except RuntimeError as err:
    print(err)
    print("This is not a linux sys")
else:
    print("Doing more linux thing")
finally:
    print("Cleaning up")
```

### Custom Exception

```python
class PlatformException(Exception):
    pass

def linu_sys():
    import sys
    if not sys.platform.startswith("linux"):
        raise PlatformException(f"Func is made for linux")
    print("Doing linux thing")

linu_sys()



```

##### raise - allow to raisse an exception
##### assert - check condition and raises an exception
##### try - all statement are executed in the block until an exception is encountered
##### except - allow to catch and handle exception that python encounter in try block
##### else - allow code to run only if there is no exception in try block
##### finally - enable to execute code that should always run, whether or not python encountered any exception
